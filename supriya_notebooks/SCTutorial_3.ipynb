```python
# Basic imports
import supriya
```

```python
note = ((supriya.ugens.LFNoise0.kr(frequency=10) + 1) * 8) + 50
supriya.graph(note.rand_range(40))
```

```python
server = supriya.Server().boot()
```

```python
@supriya.synthdef()
def sine_test(noise_hz=8):
    note = ((supriya.ugens.LFNoise0.kr(frequency=noise_hz) + 1) * 8) + 50
    freq = note.midi_to_hz()
    amp = (supriya.ugens.LFNoise1.kr(frequency=12) * .01 + 0.02)
    sig = supriya.ugens.SinOsc.ar(frequency=[freq, freq * 2]) * amp
    supriya.ugens.Out.ar(bus=0, source=sig)
server.add_synthdefs(sine_test)
```

```python
# supriya.graph(sine_test)
```

```python
synth = server.add_synth(sine_test, noise_hz=4)
```

```python
synth.free()
```

```python
server.quit()
```
